```python
N = 1000
K = 40

delta = 0.5
xi = 0.1
d = 2
max_time = 60
buffer_dim = 5000
verbose = 0

random.seed(42)
for i in range(10):

    print(f'--- Instance {i+1} ---')
    g, thetas = create_pa_graph(n_nodes=N, c=1, seed= i, init_nodes=3, init_mode="complete")  

    x0 = degree_threshold(g, N, K, thetas = thetas, connected = 1)
    s0 = connected_component_spread(g, x0, thetas=thetas, max_t=1000)[0]
    print(f'Initial spread: {s0}')
    s, x, history = Neighbor_Search(g, thetas, x0, delta, xi, d, max_time, buffer_dim, verbose = verbose)
    print(f'NS spread: {s[-1]}')
    s, x, history = NS_marta(g, x0, thetas, max_time=max_time, h_radius=30)
    print(f'NS_marta spread: {s[-1]}')
    s, x, history = NS_matteo(g, thetas, x0, delta, xi, d, max_time, buffer_dim, verbose=0, min_conn=20, mg_max_depth=10)
    print(f'NS_matteo spread: {s[-1]}')
    print('\n\n')
```
